### Importação dos dados

````python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
````

#1. Análise do faturamento


````python
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print(f"Faturamento total da Loja 1: {faturamento_loja1:.2f}")
print(f"Faturamento total da Loja 2: {faturamento_loja2:.2f}")
print(f"Faturamento total da Loja 3: {faturamento_loja3:.2f}")
print(f"Faturamento total da Loja 4: {faturamento_loja4:.2f}")
````

# 2. Vendas por Categoria


````python
vendas_categoria_loja1 = loja['Categoria do Produto'].value_counts()
vendas_categoria_loja2 = loja2['Categoria do Produto'].value_counts()
vendas_categoria_loja3 = loja3['Categoria do Produto'].value_counts()
vendas_categoria_loja4 = loja4['Categoria do Produto'].value_counts()

print("Vendas por Categoria - Loja 1:")
print(vendas_categoria_loja1)
print("\nVendas por Categoria - Loja 2:")
print(vendas_categoria_loja2)
print("\nVendas por Categoria - Loja 3:")
print(vendas_categoria_loja3)
print("\nVendas por Categoria - Loja 4:")
print(vendas_categoria_loja4)
````

# 3. Média de Avaliação das Lojas

````python
media_avaliacao_loja1 = loja['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print(f"Média de Avaliação da Loja 1: {media_avaliacao_loja1:.2f}")
print(f"Média de Avaliação da Loja 2: {media_avaliacao_loja2:.2f}")
print(f"Média de Avaliação da Loja 3: {media_avaliacao_loja3:.2f}")
print(f"Média de Avaliação da Loja 4: {media_avaliacao_loja4:.2f}")
````

# 4. Produtos Mais e Menos Vendidos

````python
produtos_mais_menos_vendidos_loja1 = loja['Produto'].value_counts()
print("Produtos Mais e Menos Vendidos - Loja 1:")
print("Mais Vendidos:")
print(produtos_mais_menos_vendidos_loja1.head())
print("\nMenos Vendidos:")
print(produtos_mais_menos_vendidos_loja1.tail())

produtos_mais_menos_vendidos_loja2 = loja2['Produto'].value_counts()
print("\nProdutos Mais e Menos Vendidos - Loja 2:")
print("Mais Vendidos:")
print(produtos_mais_menos_vendidos_loja2.head())
print("\nMenos Vendidos:")
print(produtos_mais_menos_vendidos_loja2.tail())

produtos_mais_menos_vendidos_loja3 = loja3['Produto'].value_counts()
print("\nProdutos Mais e Menos Vendidos - Loja 3:")
print("Mais Vendidos:")
print(produtos_mais_menos_vendidos_loja3.head())
print("\nMenos Vendidos:")
print(produtos_mais_menos_vendidos_loja3.tail())

produtos_mais_menos_vendidos_loja4 = loja4['Produto'].value_counts()
print("\nProdutos Mais e Menos Vendidos - Loja 4:")
print("Mais Vendidos:")
print(produtos_mais_menos_vendidos_loja4.head())
print("\nMenos Vendidos:")
print(produtos_mais_menos_vendidos_loja4.tail())
````

# 📑 Relatório de Análise de Desempenho – Alura Store



### 🎯 **Objetivo da Análise**

Prezado Senhor João,

Este relatório consolida uma análise aprofundada dos dados de desempenho das quatro lojas da sua rede Alura Store. O objetivo central é identificar, com base em evidências quantitativas, qual das unidades apresenta a menor eficiência operacional e financeira, fornecendo assim uma recomendação clara e data-driven sobre qual loja vender para viabilizar seu novo empreendimento. A decisão será fundamentada em métricas-chave, como faturamento, popularidade de produtos e categorias, satisfação do cliente e custos de frete.

### 📈 **Análise Detalhada das Métricas**

Para chegarmos a uma conclusão robusta, mergulhamos nos números de cada loja, examinando os pilares que sustentam o sucesso de um negócio de varejo.

#### 💰 **1. Faturamento Total**

O faturamento é o indicador mais direto da saúde financeira de cada loja. Ele nos mostra qual unidade está convertendo seu estoque em receita de forma mais eficaz.

```
Faturamento total de cada loja (R$)
Loja 1 - R$1.534.509,12 <---
Loja 2 - R$1.488.459,06
Loja 3 - R$1.464.025,03
Loja 4 - R$1.384.497,58
```

**Conclusão sobre o Faturamento:** Os dados mostram uma hierarquia clara. A **Loja 1** lidera em receita, enquanto a **Loja 4** figura na última posição, com uma diferença de aproximadamente **R$ 150.000,00** em relação à primeira colocada. Este é o nosso primeiro e mais forte sinal de alerta.

#### ⭐ **2. Média das Avaliações de Clientes**

A satisfação do cliente é um termômetro da reputação da marca e um forte indicador de fidelidade e sucesso a longo prazo. Uma loja com clientes insatisfeitos, mesmo que fature bem, possui um risco oculto.

```
Avaliação Média (de 1 a 5)
 Loja 1 - 3.98
 Loja 2 - 4.04
 Loja 3 - 4.05 <---
 Loja 4 - 4.00
```

**Conclusão sobre as Avaliações:** Este é um ponto extremamente relevante. Todas as quatro lojas possuem uma **avaliação média excelente**, orbitando em torno de 4 estrelas. Isso indica que a qualidade do atendimento e dos produtos é percebida de forma muito positiva e homogênea em toda a rede. Notavelmente, não há uma loja "problemática" neste quesito; pelo contrário, a satisfação do cliente é um ponto forte de todas elas.

#### 🚚 **3. Frete Médio por Loja**

O custo de frete pode ser um fator decisivo na hora da compra. Lojas com fretes mais competitivos tendem a ter uma vantagem.

```
Frete Médio (R$) |
 Loja 1 - R$ 34,69
 Loja 2 - R$ 33,62
 Loja 3 - R$ 33,07
 Loja 4 - R$ 31,28 <---
```

**Conclusão sobre o Frete:** Aqui temos um insight contraintuitivo e poderoso. A **Loja 4**, que possui o menor faturamento, é também a que oferece o **frete médio mais baixo e competitivo**. Em contrapartida, a Loja 1, líder de faturamento, tem o frete mais caro. Isso sugere que o frete não é o principal fator que impulsiona as vendas na sua rede.

#### 📦 **4. Categorias e Produtos Mais Vendidos**

Analisar o que vende mais (e menos) nos ajuda a entender o perfil de consumo e a eficiência do catálogo de cada loja.

* **Categorias Populares:** A análise mostra que todas as lojas têm um perfil de vendas muito similar, com "móveis" e "eletrônicos" liderando o ranking de unidades vendidas. Não há uma loja com um mix de categorias claramente defasado.
* **Produtos Mais/Menos Vendidos:** A distribuição de vendas entre os produtos "campeões" e os "menos populares" também é bastante parecida entre as lojas. Todas possuem seus best-sellers, com vendas entre 55 e 65 unidades, e seus produtos de menor giro, com vendas acima de 30 unidades. Isso indica uma gestão de portfólio relativamente consistente em toda a rede.

---

### 📊 **Visão Consolidada e Pontos Críticos**

Ao cruzarmos todas as informações, o cenário se torna nítido:

* As **Lojas 1, 2 e 3** formam um bloco coeso e de alta performance. Elas possuem um faturamento robusto e muito próximo entre si, além de contarem com o aval de seus clientes, refletido nas excelentes médias de avaliação.
* A **Loja 4**, embora operacionalmente eficiente em alguns aspectos (como o frete mais baixo) e com clientes satisfeitos, não consegue converter esses pontos positivos em resultado financeiro. Ela é o claro **ponto fora da curva no que tange à geração de receita**.

---

### 🏆 **Veredito Final: A Recomendação**

Senhor João,

Com base na análise aprofundada de todos os indicadores de desempenho, a recomendação é a venda da **Loja 4**.

**Justificativa Estratégica:**

1.  **Desempenho Financeiro Inferior:** A Loja 4 é a que gera o menor retorno financeiro para a rede. A diferença de faturamento para as demais é o fator mais impactante e objetivo para esta decisão. Vender o ativo de menor performance é a escolha mais lógica para otimizar seus investimentos.
2.  **Eficiência Operacional sem Conversão:** Apesar de ter o frete mais competitivo, este benefício não se traduziu em maior volume de vendas. Isso sugere que o desafio da Loja 4 pode estar relacionado a fatores mais complexos, como localização, mercado local ou marketing, que demandariam mais esforço e investimento para serem revertidos.
3.  **Saúde das Demais Lojas:** As Lojas 1, 2 e 3 são ativos valiosos e de desempenho comprovado. Elas não apenas faturam bem, como mantêm seus clientes satisfeitos, o que as torna essenciais para a saúde e o crescimento futuro da marca Alura Store. Vender qualquer uma delas seria abrir mão de uma operação claramente bem-sucedida.


A venda da Loja 4 liberará capital e, mais importante, recursos gerenciais para que o senhor possa focar no seu novo empreendimento, mantendo o núcleo mais forte e lucrativo da sua rede atual em plena operação.
